```python
# Import libraries
from pathlib import Path
import os

# Make sure notebook uses the correct project folder
os.chdir("/Users/parag/Projects/battery-health-rul-platform")

print("Current working directory:")
print(os.getcwd())

# Define dataset folder
DATA_PATH = Path("data/raw/nasa")

print("\nDoes dataset folder exist?")
print(DATA_PATH.exists())

# Find all MATLAB files
mat_files = list(DATA_PATH.glob("**/*.mat"))

print("\nNumber of .mat files found:")
print(len(mat_files))

print("\nExample .mat files:")
for file in mat_files[:10]:
    print(file)
```

```python
# Import scipy loader
import scipy.io

# Load the first .mat file
first_file = mat_files[0]

print("Opening file:")
print(first_file)

data = scipy.io.loadmat(first_file)

print("\nKeys in loaded file:")
print(data.keys())
```

```python
# Extract actual battery data
battery_key = list(data.keys())[-1]  # this will be 'B0052'

battery_data = data[battery_key]

print("Battery key:", battery_key)
print("Type:", type(battery_data))
```

```python
# Inspect structure inside battery data
print(battery_data.dtype)
```

```python
# Extract cycle data
cycles = battery_data['cycle'][0][0]

print("Number of cycles:", len(cycles))
```

```python
# Inspect first cycle
first_cycle = cycles[0]

print(type(first_cycle))
print(first_cycle.dtype)
```

```python
# Check cycle type
print("Cycle type:", first_cycle['type'])

# Check data inside cycle
cycle_data = first_cycle['data'][0][0]

print("\nAvailable signals:")
print(cycle_data.dtype)
```

```python
# Check the structure of one cycle more carefully

first_cycle = cycles[0]

print("Type of first_cycle:")
print(type(first_cycle))

print("\nFields in first_cycle:")
print(first_cycle.dtype.names)

print("\nType of first_cycle['data']:")
print(type(first_cycle["data"]))

print("\nShape of first_cycle['data']:")
print(first_cycle["data"].shape)

print("\nfirst_cycle['data']:")
print(first_cycle["data"])
```
